```
import matplotlib.pyplot as plt 
import math

print("Método de la Secante\n")

class Secante:

	__cont = 0
	__xi = []
	__ea = []

	def __init__(self, x1, x2, cp = 0.0001):		

		self.x1 = x1
		self.x2 = x2	
		self.cp = cp	
		self.__xi.append(self.x1)
		self.__xi.append(self.x2)
		self.__ea.append(1)


	def f(self, x1, x2):# x1 = Xi-1, x2 = Xi

		return x2 - ((math.exp(-x2) - x2)*(x1 - x2)) / ((math.exp(-x1) - x1)-(math.exp(-x2) - x2))	

	def resultado(self):

		while(self.__ea[self.__cont] > self.cp):

			self.__xi.append(self.f(self.__xi[self.__cont], self.__xi[self.__cont+1]))
			self.__ea.append(abs(self.__xi[self.__cont+1] - self.__xi[self.__cont]))
			self.__cont += 1
	
	def mostrarResultado(self):

		print("\n{:^0}{:^18}".format("Valor en X", "Error"))

		for i in range(len(self.__xi)):

			print("{:^11.4f}".format(self.__xi[i]), "{:^13.4f}".format(self.__ea[i-1]))

	def graficarFuncion(self):

		v = [0, 5, 0, 1]
		
		plt.plot(self.__xi, "b", label = "Función")
		plt.plot(self.__ea, "p--", label = "Error")
		plt.xlabel("Eje X")
		plt.ylabel("Eje Y")
		plt.title("Método de Newton-Raphson")
		plt.legend(loc = 1)
		#plt.text(1, 1, 'prueba')# Coloca un texto en una coordenada especificada
		plt.axis(v)# limita el rango de la gráfica especificado en el vector "v"
		plt.show()


x1 = int(input("Ingrese el primer valor para X\nXi-1 = "))
x2 = int(input("\nIngrese el segundo valor para X\nXi = "))
cp = float(input("\nIngrese un criterio de paro.\ncp = "))

miFuncion = Secante(x1, x2, cp)

miFuncion.resultado()
miFuncion.mostrarResultado()
miFuncion.graficarFuncion()

```
