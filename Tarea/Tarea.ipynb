Hacer una función que el usuario introduzca un número y la función debe de desplegar 
un triángulo de sumas, Como ejemplo, si el usuario manda a la función 4 
debeía imprimir el siguiente arreglo
+
++
+++
++++

```
def triangulo(num):
    
    i = 1

    while True:        
        for i in range(1, num+1):                                 
            for k in range(0, i):
                print("+", end="")                                    
            print()
            i = i+1                        
        break;   

triangulo(4)
```
