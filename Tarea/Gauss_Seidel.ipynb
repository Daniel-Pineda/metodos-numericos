```
import matplotlib.pyplot as plt

def fx(y):
	return 0.25+(0.25*y)

def fy(x, z):
	return 0.25+(0.25*x)+(0.25*z)

def fz(y, w):
	return 0.25+(0.25*y)+(0.25*w)

def fw(z):
	return 0.25+(0.25*z)

# DECLARACIÃ“N DE VARIABLES
listaX = []
listaY = []
listaZ = []
listaW = []
cont = 0

listaX.append(0)
listaY.append(0)
listaZ.append(0)
listaW.append(0)

while (cont < 20):
	listaX.append(fx(listaY[cont]))	
	listaY.append(fy(listaX[cont+1], listaZ[cont]))
	listaZ.append(fz(listaX[cont+1], listaW[cont]))
	listaW.append(fw(listaZ[cont+1]))
	cont+=1

for i in range(len(listaX)):	
	print("x"+str(i),"=", "{:^3.5f}, ".format(listaX[i]), end = " ")
	print("y"+str(i),"=", "{:^3.5f}, ".format(listaY[i]), end = " ")
	print("z"+str(i),"=", "{:^3.5f}, ".format(listaZ[i]), end = " ")
	print("w"+str(i),"=", "{:^3.5f}, ".format(listaW[i]), end = "\n")

plt.plot(listaX, label = "X")
plt.plot(listaY, label = "Y")
plt.plot(listaZ, label = "Z")
plt.plot(listaW, label = "W")
plt.legend()
plt.show()
```
