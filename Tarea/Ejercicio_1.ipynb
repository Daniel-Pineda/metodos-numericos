```
import matplotlib.pyplot as plt 
import math
def fx(y, z):
	return -0.22-(0.22*y)+(0.22*z)

def fy(x, z):
	return 0.375-(0.875*x)-(0.625*z)

def fz(x, y):
	return -0.6+(0.3*x)+(0.4*y)

# DECLARACIÓN DE VARIABLES
listaX = []
listaY = []
listaZ = []
cont = 0

listaX.append(0)
listaY.append(0)
listaZ.append(0)

while (cont < 10):
	listaX.append(fx(listaY[cont], listaZ[cont]))	
	listaY.append(fy(listaX[cont], listaZ[cont]))
	listaZ.append(fz(listaX[cont], listaY[cont]))	
	cont+=1

for i in range(len(listaX)):	
	print("x"+str(i),"=", "{:^3.5f}, ".format(listaX[i]), end = " ")
	print("y"+str(i),"=", "{:^3.5f}, ".format(listaY[i]), end = " ")
	print("z"+str(i),"=", "{:^3.5f}, ".format(listaZ[i]), end = "\n")	

plt.plot(listaX, label = "X")
plt.plot(listaY, label = "Y")
plt.plot(listaZ, label = "Z")
plt.legend()
plt.show()
```
