Funciones

```
import matplotlib.pyplot as plt
import math 

def g1(x):
	
	return 2*x**2-5

def g2(x, cp = 0.0001):

	return 5/(2*x-1)

def g3(x, cp = 0.0001):
	
	return math.sqrt((x+5)/2)
```

Iteración de las funciones

```
listaG1 = []
errorG1 = []
listaG2 = []
errorG2 = []
listaG3 = []
errorG3 = []
cont = 0
cp = 0.0001


x = int(input("Establece el valor de x: "))

listaG1.append(x)
listaG2.append(x)
listaG3.append(x)

errorG1.append(1)
errorG2.append(1)
errorG3.append(1)

print("g1(x)")

while (errorG1[cont] > cp):
	
	listaG1.append(g1(listaG1[cont]))
	errorG1.append(abs(listaG1[cont+1] - listaG1[cont]))	

	cont += 1

	if (cont == 5 and errorG1[cont] > 1):

		print("No converge")
		break;

print(listaG1)
print(errorG1)

print("g2(x)")

cont = 0

while (errorG2[cont] > cp):
	
	listaG2.append(g2(listaG2[cont]))
	errorG2.append(abs(listaG2[cont+1] - listaG2[cont]))	

	cont += 1

	if (cont == 8 and errorG2[cont] > 1):

		print("No converge")
		break;	
	
	if (cont == 8 and errorG2[cont] < 1):

		print("Sí converge")
		break;	

print(listaG2)
print(errorG2)

print("g3(x)")

cont = 0

while (errorG3[cont] > cp):
	
	listaG3.append(g3(listaG3[cont]))
	errorG3.append(abs(listaG3[cont+1] - listaG3[cont]))	

	cont += 1

	if (cont == 5 and errorG3[cont] > 1):

		print("No converge")
		break;	
	
	if (cont == 5 and errorG3[cont] < 1):

		print("Sí converge")
		break;	

print(listaG3)
print(errorG3)

plt.plot(listaG1)
plt.show()

plt.plot(listaG2)
plt.show()

plt.plot(listaG3)
plt.show()
```
